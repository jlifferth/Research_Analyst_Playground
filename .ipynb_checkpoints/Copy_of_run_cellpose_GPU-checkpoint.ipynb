```python
import subprocess
import datetime as dt

```

```python
dt_string = dt.datetime.now().strftime("%d-%m-%Y_%H:%M:%S")
normal_filename = 'profiles/{}profile_{}.txt'.format('normal_', dt_string)
fast_filename = 'profiles/{}profile_{}.txt'.format('fast_', dt_string)
print(normal_filename) 
print(fast_filename)
```

```python
normal_command = 'python -m cProfile -m cellpose --dir /content/lung/ --pretrained_model nuclei --save_tif --channel_axis 0 --chan 1 --verbose --use_gpu > gpu_normal_profile_18-02-2022_14:56:57.txt{}'.format(normal_filename)
fast_command = 'python -m cProfile -m cellpose --dir /content/lung/ --pretrained_model nuclei --save_tif --channel_axis 0 --chan 1 --verbose --fast_mode > {}'.format(fast_filename)
subprocess.run(normal_command, shell=True)
```
