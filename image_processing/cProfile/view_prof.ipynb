```python
import os
prof_paths = []
target_dir = 'profiles/'

for root, dirs, files in os.walk(target_dir):
    for file in files:
        if file.endswith('.csv'):
            prof_paths.append(os.path.join(root,file))

prof_paths.sort()
print(prof_paths)

dfs = []

for path in prof_paths:
    df = pd.read_csv(path)
    dfs.append(df)

print(type(dfs))
```

```python
df = dfs[0]
df = df.sort_values(by=['tottime'], ascending=False)
df = df.reset_index(drop=True)
df
```
